# 1. What is Plotly ?

The Plotly Python library is an interactive open-source library. This can be a very helpful tool for data visualization and understanding the data simply and easily. plotly graph objects are a high-level interface to plotly which are easy to use. It can plot various types of graphs and charts like scatter plots, line charts, bar charts, box plots, histograms, pie charts, etc.

# 2. Where is it used ?

Plotly is used mainly in Scatterplots, Bar Charts, Pie Charts, Histograms, Box Plots.

Scatter Plot:-  Scatter plot represent values for two different numeric variables. They are mainly used for representation of relationship between two variables.

Bar charts:-  Bar charts are used when we want to compare different groups of data and make inference of which groups are highest and which groups are comon and compare how one group is performing compared to others.

Pie chart:-   A pie chart represents the distribution of different variables among total. In pie chart each slice shows itâ€™s contribution to the total amount.

Histogram:-    A histogram plots the continuous distribution of variable as series of bars and each bar indicates the frequency of the occurring value in a variable. In order to use a histogram, we simply require a variable that takes continuous numeric values.

Box plot:-  A box plot is the representation of statistical summary. Minimum, First Quartile, Median, Third Quartile, Maximum.



# 3. plot a Pie Chart of Indian Car companies market share % values: max ( 6 Companies market share )

```python
import numpy as np
import matplotlib.pyplot as plt

labels = 'Maruti Suzuki', 'Hyundai Motor India', 'Mahindra & Mahindra','Tata Motors','Honda Cars India','Toyota Kirloskar','Renault India', 'Ford India'
  
sizes = [1643467, 536241, 248859, 210200, 170026, 140645, 102222, 90061]
colors = ['red', 'blue', 'yellow', 'green', 'pink', 'grey', 'maroon', 'orange']
explode = (0.3, 0.21, 0.1, 0.21, 0.21, 0.2, 0.25, 0.28)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')
plt.show()
```

# 4. Plot a Bar Chart with India's covid-19 Active cases ( Max 10 States )

Maharastra = MH, Kerala = KL, Karnataka = KA,Tamil Nadu = TN, Andhra Pradesh = AP, Uttar Pradesh = UP, West Bengal = WB, Delhi = DL, Chattisgarh = CG, Rajasthan = RJ, Odisha = OD, Gujarat = GJ, 

States = https://en.wikipedia.org/wiki/List_of_Regional_Transport_Office_districts_in_India

Active cases = https://www.statista.com/statistics/1103458/india-novel-coronavirus-covid-19-cases-by-state/



```python
import matplotlib.pyplot as plt
   
States = ['MH','KL','KA','TN', 'AP', 'UP', 'WB', 'DL','CG', 'RJ', 'OD', 'GJ']
        
Activecases = [116165, 114849, 36737, 32307, 28680, 1594, 14901, 743, 4784, 702, 22637, 931]

colors = ['red','blue','green', 'grey', 'pink', 'lightblue', 'orange', 'purple', 'yellow', 'violet', 'lightblue', 'lightgreen']

plt.bar(States, Activecases, color=colors)
plt.title('States Vs Activecases', fontsize=15, )
plt.xlabel('States', fontsize=10)
plt.ylabel('Activecases', fontsize=10)
plt.grid(True)
plt.show()
```
