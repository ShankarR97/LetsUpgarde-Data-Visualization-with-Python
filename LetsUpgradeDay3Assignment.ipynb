# Plot using sns.relplot, line plot, box plot on fmri data set


```python
import numpy as np
import seaborn as sns
sns.set_style('darkgrid')
```

```python
data= sns.load_dataset('fmri')
data.head()
```

**Regression Plot

```python
sns.relplot(x = 'timepoint', y = 'signal', data = data , hue = 'region',aspect=3, palette="mako")
```

```python
sns.relplot(x = 'timepoint', y = 'signal', data = data , hue = 'event',aspect=3, palette="PuRd")
```

```python
sns.relplot(x="timepoint", y="signal", hue="region", kind="line", data=data,aspect=3, palette="Purples")
```

```python
sns.relplot(x="timepoint", y="signal", hue="region", kind="line", data=data,aspect=3, palette="Wistia")
```

```python
sns.boxplot(x="timepoint", y="signal", hue="event", data=data, fliersize=4, palette="YlGn")
```

```python
sns.boxplot(x="timepoint", y="signal", hue="event", data=data, fliersize=4, palette="YlOrBr")
```
